```python
def sendMail(email,name):                   #To generate a funtion for general mailing ease
    
    html_text = '''<p>Hey candidate('''+ name +'''),</p>
<p>We are pleased to offer you a job as a Social Media Marketer at HR Solutions.</p>
<p>We think that your experience and skills will be a valuable asset to our company.</p>
<p>&nbsp;If you are to accept this offer you will be eligible to the following in accordance to our company&apos;s policies.</p>'''

    subject = "Hello candidate-"+ name +",here is your offer letter"                                  #Add a subject with name 
    message = emails.html(html= html_text,                                             #Adding Text to Html
                          subject=subject,                                             #Subject location name
                          mail_from=('Rema Mallick','Rema@hrsolutions.com'))           #Sender Details

    mail_via_python = message.send(to=email, 
                               smtp={'host': 'smtp.gmail.com',            #SMTP for gmail
                                     'timeout': 5,
                                     'port':587,
                                     'user':'adityajama@serverinc.com',                        #please insert your email in user, this is SMTP email id
                                     'password':'Adityajama2234',        #please insert your email-password here
                                     'tls':True})         
    return mail_via_python.status_code
```
